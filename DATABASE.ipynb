```python
# Load the CSV file into a DataFrame
df = pd.read_csv("Bus_Routes.csv")

# Add a new 'id' column with default values (e.g., incrementing numbers)
df.insert(0, 'id', range(1, len(df) + 1))

# Rename columns with the new 'id' column included
df.columns = ['id', 'route_Name', 'route_Link', 'bus_Name', 'bus_types', 'departing',
              'duration', 'arrival', 'rating', 'price', 'seats']

# Print DataFrame to verify the new column names
print("\nAfter renaming and adding 'id':")
print(df.head())
print(df.columns)

# Save the DataFrame with the new column names to a new CSV file (optional)
df.to_csv('BusRoutes.csv',index=False)
```

```python
insert = """insert into  buses(id, route_Name, route_Link, bus_Name, bus_types, departing,
       duration, arrival, rating, price, seats) values(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"""
```

```python
from tqdm import tqdm

for index, row in tqdm (df.iterrows()):
    cursor.execute(insert, (row["id"],row["route_Name"], row["route_Link"], row["bus_Name"], row["bus_types"],row["departing"],row["duration"],row["arrival"],row["rating"],row["price"],row["seats"]))
    conn.commit()
```
