```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import time

# Setup driver
driver = webdriver.Chrome()
wait = WebDriverWait(driver, 10)

# Initialize empty lists
Route_Name = []
Route_Links = []

# Open the redbus URL
driver.get('https://www.redbus.in/online-booking/ksrtc-kerala/?utm_source=rtchometile')
driver.maximize_window()

# Wait until the data loads
wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "route")))

# Function to extract text and links
def getText():
    elements = driver.find_elements(By.CLASS_NAME, "route")
    for element in elements:
        # You might need to adjust this if there are no 'href' attributes
        link = element.get_attribute("href") if element.get_attribute("href") else "No link"
        Route_Name.append(element.text)
        Route_Links.append(link)

# Initial extraction
getText()

# Loop through pages to scrape data
for page_num in range(1, 2):  # Changed range to start from 1 to match typical page numbers
    try:
        if page_num <= 2:  # Ensure page_num is within the expected range
            # Wait for pagination container and locate next page button
            pagination_container = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[12]')))
            next_page_button = pagination_container.find_element(By.XPATH, f'.//div[contains(@class,"DC_117_pageTabs") and text()="{page_num+1}"]')

            # Scroll to and click the next page button
            actions = ActionChains(driver)
            actions.move_to_element(next_page_button).perform()
            time.sleep(2)
            driver.execute_script("arguments[0].scrollIntoView();", next_page_button)
            time.sleep(2)
            next_page_button.click()
            time.sleep(5) 
            print(f"Successfully navigated to page {page_num}")
            getText()  # Extract data from the new page
        else:
            print(f"Page number {page_num} is greater than 2, skipping...")
    except Exception as e:
        print(f"An error occurred on page {page_num}: {e}")

# Print the results
print("Route Names are:\n", "\n".join(Route_Name))
print("Route Links are:\n", "\n".join(Route_Links))

# Close the driver
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Initialize the driver
driver = webdriver.Chrome()
wait = WebDriverWait(driver, 10)

# Empty Datasets
Bus_names = []
Bus_types = []
Depart_times = []
Durations = []
Arrivals = []
Ratings = []
prices = []
seats_avail = []

# Loop through Route_Links to scrape data
for link in Route_Links:
    driver.get(link)
    time.sleep(2)  # Consider using WebDriverWait instead of time.sleep()

    # Click element on bus
    click_elements = driver.find_elements(By.XPATH, f"//a[contains(@href,'{link}')]")
    for click_element in click_elements:
        driver.execute_script("arguments[0].scrollIntoView(true);", click_element)
        time.sleep(1)
        try:
            click_element.click()
        except Exception as e:
            print(f"Exception while clicking: {e}")
            driver.execute_script("arguments[0].click();", click_element)
            time.sleep(2)

    # Click to view bus details
    view_bus = driver.find_elements(By.XPATH, "//div[@class='button']")
    for element_click in view_bus:
        driver.execute_script("arguments[0].scrollIntoView(true);", element_click)
        time.sleep(2)
        try:
            element_click.click()
        except Exception as e:
            print(f"Exception while clicking: {e}")
            driver.execute_script("arguments[0].scrollIntoView();", element_click)
            time.sleep(2)

    # Scrolling to load all elements
    scrolling = True
    while scrolling:
        old_page_source = driver.page_source
        print("Old page source captured")

        body = driver.find_element(By.TAG_NAME, "body")
        body.send_keys(Keys.PAGE_DOWN)
        time.sleep(2)  # Adjust sleep time as needed
        new_page_source = driver.page_source
        if new_page_source == old_page_source:
            scrolling = False

    wait = WebDriverWait(driver, 50)
    try:
        wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")))
        wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']")))
        wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[@class='dp-time f-19 d-color f-bold']")))
        wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[@class='dur l-color lh-24']")))
        wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[@class='bp-time f-19 d-color disp-Inline']")))
        wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[@class='rating-sec lh-24']")))
        wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[@class='fare d-block']")))
        wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[@class='seat-left m-top-30']")))

        # Extracting the data
        Bus_name_elements = driver.find_elements(By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")
        Bus_type_elements = driver.find_elements(By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']")
        Depart_time_elements = driver.find_elements(By.XPATH, "//div[@class='dp-time f-19 d-color f-bold']")
        Duration_elements = driver.find_elements(By.XPATH, "//div[@class='dur l-color lh-24']")
        Arrival_elements = driver.find_elements(By.XPATH, "//div[@class='bp-time f-19 d-color disp-Inline']")
        Rating_elements = driver.find_elements(By.XPATH, "//div[@class='rating-sec lh-24']")
        price_elements = driver.find_elements(By.XPATH, "//div[@class='fare d-block']")
        Seats_elements = driver.find_elements(By.XPATH, "//div[@class='seat-left m-top-30']")

        Bus_names.extend([bus.text for bus in Bus_name_elements])
        Bus_types.extend([types.text for types in Bus_type_elements])
        Depart_times.extend([Depart.text for Depart in Depart_time_elements])
        Durations.extend([time.text for time in Duration_elements])
        Arrivals.extend([endtime.text for endtime in Arrival_elements])
        Ratings.extend([star.text for star in Rating_elements])
        prices.extend([rs.text for rs in price_elements])
        seats_avail.extend([chair.text for chair in Seats_elements])

    except Exception as e:
        print(f"An error occurred: {e}")

# Print the data
print("Bus Names are:\n", "\n".join(Bus_names))
print("Bus types are:\n", "\n".join(Bus_types))
print("Departings are:\n", "\n".join(Depart_times))
print("Durations are:\n", "\n".join(Durations))
print("Arrivals are:\n", "\n".join(Arrivals))
print("Ratings are:\n", "\n".join(Ratings))
print("Prices are:\n", "\n".join(prices))
print("Seats available are:\n", "\n".join(seats_avail))

# Quit the driver
driver.quit()
```
